### visualizing multiple regression with `lmplot()`

input data must be in a pandas `DataFrame`

to plot provide the `predicator` and `response variables` names along width the dataset

```python
tips = sns.load_dataset('tips')
```

```python
sns.lmplot('total_bill', 'tip', tips)
```

this plot has two main components

1. scatterplot => showing the observed datapoints

1. regression line => showing the estimated linear model relating the two variables

```python
# `ci` confidence interval
sns.lmplot('total_bill', 'tip', tips, ci=68)
```

```python
# control the aesthetics of the underlying plots
sns.lmplot('total_bill', 'tip', tips,
           scatter_kws={'marker': '.', 'color': 'slategrey'},
           line_kws={'linewidth': 1, 'color': 'seagreen'})
```

### plotting with discrete predictor variables

independent variable is quantitative but discrete

```python
sns.lmplot('size', 'tip', tips)
```

```python
# improved with a bit of jitter
sns.lmplot('size', 'tip', tips, x_jitter=.15)
```

```python
# estimate the central tendency of each bin
# `x_estimator`
# pass function that aggregates a vector of data into one estimate
sns.lmplot('size', 'tip', tips, x_estimator=np.mean)
```

```python
# plot an estimated central tendency and confidence interval
# provide specific centroid values for the bins
bins = [10, 20, 30, 40]
sns.lmplot('total_bill', 'tip', tips, x_bins=bins)
```

```python
# give a number of bins
# it will find centroids
sns.lmplot('total_bill', 'tip', tips, x_bins=8)
```

### faceted linear model plots

`lmplot()` function is built on top of a `FacetGrid`

```python
# using a hue facet makes it easiest to directly compare the two subsets
sns.lmplot('total_bill', 'tip', tips, hue='smoker')
```

```python
# scatterplot marker
sns.lmplot('total_bill', 'tip', tips, hue='smoker', markers=['x', 'o'])
```

```python
# plotting in different columns
sns.lmplot('total_bill', 'tip', tips, col='smoker')
```

```python
# it returns the grid object
g = sns.lmplot('total_bill', 'tip', tips, hue='day', palette='Set2',
               hue_order=['Thur', 'Fri', 'Sat', 'Sun'])
g.set_axis_labels('total bill (us dollars)', 'tips')
g.set(xticks=[10, 30, 50], ylim=(0, 10), yticks=[0, 2.5, 5, 7.5, 10])
```

### ploting different linear relationships

default => lmplot() shows the ordinary least squares fit to the data

```python
sns.lmplot('total_bill', 'tip', tips, hue='time', palette='Set1', fit_reg=False)
```

```python
# plotting nonlinear trends
sns.lmplot('size', 'total_bill', tips, order=2)
```

```python
# plot the lowess line
sns.lmplot('total_bill', 'tip', tips, lowess=True, line_kws={'color': '.2'})
```

```python
# plotting logistic regression
# define a dependent measure that takes the values `True` and `False`
tips['big_tip'] = (tips['tip'] / tips['total_bill']) > .15
```

```python
# plot these data as normal
sns.lmplot('total_bill', 'big_tip', tips)
```

```python
# this plot suggests that diners with a larger total bill
# are less likely to leave a big tip
# adding a bit of jitter to the scatter plot
sns.lmplot('total_bill', 'big_tip', tips, y_jitter=.05)
```

```python
# fit a logistic regression and plot the resulting curve over the data
sns.lmplot('total_bill', 'big_tip', tips, y_jitter=.05, logistic=True)
```

```python
# plotting data with outliers
# linear regression is based on the mean
# it can be overly influenced by outliers
# robust estimation is computationally intensive
sns.lmplot('total_bill', 'tip', tips, robust=True, n_boot=500)
```

### controlling for confounding variables

```python
c = np.random.randn(50)
a = 2 + c + np.random.randn(50)
b = 3 + c + np.random.randn(50)
df = pd.DataFrame(dict(A=a, B=b, C=c), columns=list('ABC'))
```

```python
# plot the relationship between A and B
# they will look related
sns.lmplot('A', 'B', df)
```

```python
# if you remove (or `partial out`) variable C from variable A before plotting
# you can see that they are in fact independent
sns.lmplot('A', 'B', df, x_partial='C')
```

### plotting simple regression with `regplot()`

`lmplot()` is just using a lower-level function
`regplot()` to draw the data and the regression line

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
sns.regplot('total_bill', 'tip', tips, ax=ax1)
sns.boxplot(tips['tip'], tips['size'], color='Blues_r', ax=ax2).set_ylabel('')
f.tight_layout()
```

```python
# because `lmplot()` is using `regplot()` behind the scenes
# all of the options for
#  ->  binning the datapoints
#  ->  fitting the regression
#  ->  modifying the aesthetics
# are shared
#
# what's missing form `regplot()` is the faceting options
#
# `regplot` also accepts data
#  -> numpy arrays
#  -> pandas series objects
x, y = np.random.multivariate_normal([1, 5], [(2, -.8), (-.8, 2)], 80).T
ax = sns.regplot(x, y, color='seagreen')
ax.set(xlabel='x variable', ylabel='y variable')
```

### examining model residuals using `residplot()`

it's easy to visualize the residuals of a simple model using the `residplot()`

it looks a lot like `regplot`

but instead of plotting the observations and regression model

it plots the residuals

```python
sns.residplot(x, y)
```

```python
# higher-order relationship
y = x + 1.5 * x ** 2 + np.random.randn(len(x))
sns.residplot(x, y, color='indianred')
```

```python
# visualize it by fitting a lowess smoother to the residuals
sns.residplot(x, y, color='indianred', lowess=True)
```

```python
# residplot() let you calculate residuals with a polynomial regression
# fitting a second-order regression removes this structure
sns.residplot(x, y, color='indianred', order=2, lowess=True)
```

### describing continuous interactions with `interactplot()`

```python
# two-way interaction
rs = np.random.RandomState(1)
x1, x2 = rs.normal(size=(2, 100))
y1 = .5 * x1 + 2 * x2 + 2 * x1 * x2 + rs.normal(size=100)
y2 = x1 + x2 + rs.normal(size=100)
p = 1 / (1 + np.exp(-y1))
y_flip = [rs.binomial(1, p_i) for p_i in p]
df = pd.DataFrame(dict(x1=x1, x2=x2, y1=y1, y2=y2, y_flip=y_flip))
```

```python
bins = np.array([-1, 0, 1])
binned = np.digitize(x2, bins)
df['x2_bin'] = binned
sns.lmplot('x1', 'y1', df, col='x2_bin', hue='x2_bin', palette='PuBuGn_d', size=3)
```

```python
sns.interactplot(x1, x2, y1)
```

```python
# pass a dataframe and adjust aesthetics of the plot
sns.interactplot('x1', 'x2', 'y1', df, cmap='coolwarm', filled=True, levels=25)
```

```python
# underlying plot functions
#  -> `contourf()`
#  -> `plot()`
sns.interactplot('x1', 'x2', 'y2', df, filled=True,
                 scatter_kws={'color': 'dimgray'},
                 contour_kws={'alpha': .5})
```

```python
# for logistic regression
pal = sns.blend_palette(['#4169E1', '#DFAAEF', '#E16941'], as_cmap=True)
levels = np.linspace(0, 1, 11)
sns.interactplot('x1', 'x2', 'y_flip', df, levels=levels, cmap=pal, logistic=True)
```
