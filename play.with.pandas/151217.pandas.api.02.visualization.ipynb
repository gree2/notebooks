```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2015', periods=1000))
ts = ts.cumsum()
ts.plot();
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
plt.figure(); df.plot();
```

```python
# plot one column versus another using x y keywords
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = pd.Series(list(range(len(df))))
df3.plot(x='A', y='B');
```

```python
# box plots
df = pd.DataFrame(np.random.rand(10, 5), columns=list('ABCDE'))
df.plot(kind='box');
```

```python
color = dict(boxes='DarkGreen', whiskers='DarkOrange',
            medians='DarkBlue', caps='Gray')
df.plot(kind='box', color=color, sym='r+');
```

```python
df = pd.DataFrame(np.random.rand(10, 5))
plt.figure();
bp = df.boxplot();
```

```python
df = pd.DataFrame(np.random.rand(10, 2), columns=['Col1', 'Col2'])
df['X'] = pd.Series(list('AAAAABBBBB'))
plt.figure();
bp = df.boxplot(by='X');
```

```python
df = pd.DataFrame(np.random.rand(10, 3), columns=['Col1', 'Col2', 'Col3'])
df['X'] = pd.Series(list('AAAAABBBBB'))
df['Y'] = pd.Series(list('ABABABABAB'))
plt.figure();
bp = df.boxplot(column=['Col1', 'Col2'], by=['X', 'Y']);
```

```python
np.random.seed(1234)
df_box = pd.DataFrame(np.random.randn(50, 2))
df_box['g'] = np.random.choice(['A', 'B'], size=50)
df_box.loc[df_box['g'] == 'B', 1] += 3
bp = df_box.boxplot(by='g');
```

```python
bp = df_box.groupby('g').boxplot();
```

```python
# scatter plot
df = pd.DataFrame(np.random.rand(50, 4), columns=list('abcd'))
df.plot(kind='scatter', x='a', y='b');
```

```python
# plot multiple column groups in a single axes
ax = df.plot(kind='scatter', x='a', y='b', color='DarkBlue', label='Group 1');
```

```python
ax = df.plot(kind='scatter', x='a', y='b', color='DarkBlue', label='Group 1');
df.plot(kind='scatter', x='c', y='d', color='DarkGreen', label='Group2', ax=ax);
```

```python
# keyword c may be given as the name of a column to provide colors for each point
df.plot(kind='scatter', x='a', y='b', c='c', s=50);
```

```python
df.plot(kind='scatter', x='a', y='b', s=df['c']*200);
```

```python
# hexagonal bin plot
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df.plot(kind='hexbin', x='a', y='b', gridsize=25);
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df['b'] = df['b'] + np.arange(1000)
df['z'] = np.random.uniform(0, 3, 1000)
df.plot(kind='hexbin', x='a', y='b', C='z', reduce_C_function=np.max, gridsize=25);
```

```python
# pie plot
series = pd.Series(3 * np.random.rand(4), index=list('abcd'), name='series')
series.plot(kind='pie', figsize=(6, 6));
```

```python
df = pd.DataFrame(3 * np.random.rand(4, 2), index=list('abcd'), columns=['x', 'y'])
df.plot(kind='pie', subplots=True, figsize=(8, 4));
```
