# controlling figure aesthetics

```python
%matplotlib inline

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# 1069 is sum(map(ord, 'aesthetics))
np.random.seed(sum(map(ord, 'aesthetics')))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
import seaborn as sns
sinplot()
```

use `axes_style()` and `set_style()` function to control style

use `plotting_context()` and `set_context()` function to scale the plot

## styling figures with axes_style() and set_style()

preset seaborn themes

1. darkgrid => default

1. whitegrid

1. dark

1. white

1. ticks

```python
sns.set_style('whitegrid')
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data)
```

```python
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set_style('ticks')
sinplot()
```

## removing spines with despine()

both `white` and `ticks` styles can benefit from

removing the top and right axes splines

through the matplotlib parameters

call `despine()`

```python
sinplot()
sns.despine()
```

```python
# offsetting the spines away form the data
# `trim` will limit the range of the surviving spines
f, ax = plt.subplots()
sns.violinplot(data)
sns.despine(offset=10, trim=True)
```

```python
# control which spines are removed
sns.set_style('whitegrid')
sns.boxplot(data, color='deep')
sns.despine(left=True)
```

## temporarity setting figure style

use `axes_style()` function in a `with` statement

to temporarily set plot parameters

```python
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
    
plt.subplot(212)
sinplot(-1)
```

## overriding element of the seaborn styles

pass a dictionary of parameters to the `rc` argument

```python
# get parameters
sns.axes_style()
```

```python
# then set different versions of these parameters
sns.set_style('darkgrid', {'grid.linewidth': .5,
                           'axes.facecolor': '.9'})
sinplot()
```

## scaling plot elements with plotting_context() and set_context()

1. paper

1. notebook => default

1. talk

1. poster

```python
# reset the default parameters
sns.set()
```

```python
sns.set_context('paper')
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context('talk')
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context('poster')
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context('notebook', font_scale=1.5, rc={'line.linewidth': 2.5})
sinplot()
```
